```python
#Import python libraries
import pandas as pd
import numpy as np
import os
```

OG 2003 Update

```python
#Import excel sheets from Consolidated FDI Data/2003 - 2021 FDI Africa.xlsx
OG_2003_Update = pd.read_excel(r'Consolidated FDI Data/2003 - 2021 FDI Africa.xlsx', sheet_name = 'OG 2003 Update', header = 0,engine="openpyxl",usecols="A:Z")
```

```python
#Drop rows where value in num is not an integer & reset index
OG_2003_Update = OG_2003_Update[OG_2003_Update['num'].apply(lambda x: isinstance(x, int))]
OG_2003_Update = OG_2003_Update.reset_index(drop=True)
```

OG 2018 Downloaded

```python
#Import excel sheets from Consolidated FDI Data/2003 - 2021 FDI Africa.xlsx
OG_2018_Downloaded = pd.read_excel(r'Consolidated FDI Data/2003 - 2021 FDI Africa.xlsx', sheet_name = 'OG 2018 Downloaded', header = 0,engine="openpyxl")
```

OG 2018 Scrapped

```python
#Import excel sheets from Consolidated FDI Data/2003 - 2021 FDI Africa.xlsx
OG_2018_Scrapped = pd.read_excel(r'Consolidated FDI Data/2003 - 2021 FDI Africa.xlsx', sheet_name = 'OG 2018 Scrapped', header = 0,engine="openpyxl",usecols="A:O") 
#Remove Unnamed: 0 column
OG_2018_Scrapped = OG_2018_Scrapped.drop(columns=['Unnamed: 0'])
```

Data From2003

```python
#Import excel sheets from Consolidated FDI Data/2003 - 2021 FDI Africa.xlsx and only certain columns
Data_From2003 = pd.read_excel(r'Consolidated FDI Data/2003 - 2021 FDI Africa.xlsx', sheet_name = 'Data From2003', header = 0,engine="openpyxl",usecols = "A:M")
```

Consolidated Africa FDI

```python
#Import excel sheets from Consolidated FDI Data/2003 - 2021 FDI Africa.xlsx
Consolidated_Africa_FDI = pd.read_excel(r'Consolidated FDI Data/2003 - 2021 FDI Africa.xlsx', sheet_name = 'Consolidated Africa FDI', header = 0,engine="openpyxl")
```

```python
#Print the shape of the dataframes
print(OG_2003_Update.shape)
print(OG_2018_Downloaded.shape)
print(OG_2018_Scrapped.shape)
print(Consolidated_Africa_FDI.shape)

#Sum of all rows of all dataframes
print(OG_2003_Update.shape[0] + OG_2018_Downloaded.shape[0] + OG_2018_Scrapped.shape[0] + Consolidated_Africa_FDI.shape[0])
```

```python
OG_2018_Downloaded.drop(columns=['destination_region.1'], inplace=True)
OG_2003_Update.drop(columns=['dest_country.1'], inplace=True)
```

```python
# Rename sub-sector column to subsector
OG_2018_Downloaded.rename(columns={'sub-sector':'subsector'}, inplace=True)
OG_2003_Update.rename(columns={'dest_country':'destination_country'}, inplace=True)
OG_2003_Update.rename(columns={'dest_region':'destination_region'}, inplace=True)
OG_2003_Update.rename(columns={'dest_city':'destination_city'}, inplace=True)
OG_2003_Update.rename(columns={'dest_state':'destination_state'}, inplace=True)
Consolidated_Africa_FDI.rename(columns={'dest_country':'destination_country'}, inplace=True)
OG_2018_Scrapped.rename(columns={'Sector':'sector'}, inplace=True)
OG_2018_Downloaded.rename(columns={'Date':'date'}, inplace=True)
OG_2018_Scrapped.rename(columns={'Date':'date'}, inplace=True)
OG_2018_Scrapped.rename(columns={'jobs':'jobs_created'}, inplace=True)
```

```python
#Print column names of all dataframes in ascending order
print(OG_2003_Update.columns.sort_values())
print(OG_2018_Downloaded.columns.sort_values())
print(OG_2018_Scrapped.columns.sort_values())
print(Consolidated_Africa_FDI.columns.sort_values())
```

```python
#Concatinate all dataframes column wise
Consolidated_Africa_FDI_combined = pd.concat([OG_2003_Update,OG_2018_Downloaded,OG_2018_Scrapped,Consolidated_Africa_FDI], axis = 0).reset_index(drop=True)
```
