```python
#Import python libraries
import pandas as pd
import numpy as np
import os
from datetime import datetime
#ignore warning
import warnings
warnings.filterwarnings('ignore')
```

Read all CSVs from different folders that are combined

```python
Consolidated_Africa_FDI_combined = pd.read_csv('Consolidated FDI Data/Consolidated_Africa_FDI_combined.csv')
developed_03_20 = pd.read_csv('developed/developed_03_20.csv')
developing = pd.read_csv('developing/developing.csv')
no_india_china = pd.read_excel('Projects in all developing countries no india or china through march 2018/1.xlsx',engine="openpyxl")
#Drop 1st column
no_india_china = no_india_china.drop(no_india_china.columns[0], axis=1)
```

```python
Consolidated_Africa_FDI_combined['date'] = pd.to_datetime(Consolidated_Africa_FDI_combined['date'])
Consolidated_Africa_FDI_combined['date']
```

```python
#2003-01-01 00:00:00+00:00 to 2003-01-01
developed_03_20['project_date'] = pd.to_datetime(developed_03_20['project_date'])
developed_03_20['project_date'] = developed_03_20['project_date'].dt.date
developed_03_20['project_date']
```

```python
developing['project date'][23502:] = '20' + developing['project date'][23502:].astype(str)
developing['project date'][23502:] = developing['project date'][23502:].apply(lambda x: datetime.strptime(x, "%Y-%b").strftime('%Y-%m'))
developing['project date'][23502:] = developing['project date'][23502:] + '-01'
developing['project date'][:23502] = pd.to_datetime(developing['project date'][:23502]).dt.date
developing['project date'] = pd.to_datetime(developing['project date'])
developing['project date']
```

```python
#keep alphabets and numbers in columns in lower case
Consolidated_Africa_FDI_combined.columns = Consolidated_Africa_FDI_combined.columns.str.lower()
developed_03_20.columns = developed_03_20.columns.str.lower()
developing.columns = developing.columns.str.lower()
no_india_china.columns = no_india_china.columns.str.lower()

Consolidated_Africa_FDI_combined.columns = Consolidated_Africa_FDI_combined.columns.str.replace('[^A-Za-z0-9]+', ' ')
developed_03_20.columns = developed_03_20.columns.str.replace('[^A-Za-z0-9]+', ' ')
developing.columns = developing.columns.str.replace('[^A-Za-z0-9]+', ' ')
no_india_china.columns = no_india_china.columns.str.replace('[^A-Za-z0-9]+', ' ')

developed_03_20.drop('freezone', axis=1, inplace=True)
Consolidated_Africa_FDI_combined.drop('freezone', axis=1, inplace=True)
#remove duplicate columns in developed_03_20
developed_03_20 = developed_03_20.loc[:,~developed_03_20.columns.duplicated()]
```

```python
#Rename column date to project date in Consolidated
Consolidated_Africa_FDI_combined.rename(columns={'date':'project date'}, inplace=True)
Consolidated_Africa_FDI_combined.rename(columns={'destination adm region':'destination admin region'}, inplace=True)
developed_03_20.rename(columns={'destination adm region':'destination admin region'}, inplace=True)
developing.drop('destination admin region', axis=1, inplace=True)
developing.rename(columns={'destination adm region':'destination admin region'}, inplace=True)
developing.rename(columns={'motive description':'motive'}, inplace=True)
no_india_china.rename(columns={'motive description':'motive'}, inplace=True)
developed_03_20.drop('date', axis=1, inplace=True)
developed_03_20.rename(columns={'source adm region':'source admin region'}, inplace=True)
Consolidated_Africa_FDI_combined.rename(columns={'source adm region':'source admin region'}, inplace=True)
developing.drop('source adm region', axis=1, inplace=True)
developed_03_20['project date'] = pd.to_datetime(developed_03_20['project date'])
```

```python
#Concatenate all dataframes into one column wise
df = pd.concat([Consolidated_Africa_FDI_combined,developed_03_20,developing],axis=0)
df.reset_index(drop=True, inplace=True)
#remove string in brackets in column destination city
df['destination city'] = df['destination city'].str.replace(r"\(.*\)","")
```

```python
#Print project date column of all dataframes
print(Consolidated_Africa_FDI_combined['project date'])
print(developed_03_20['project date'])
print(developing['project date'])
```

```python
lats_long = pd.read_csv('worldcities.csv')
lats_long = lats_long[['city_ascii','country','lat','lng']]
#Remove duplicate rows as per city_ascii in lats_long
lats_long = lats_long.drop_duplicates(subset=['city_ascii'], keep='first')
```

```python
#Merge lats_long and df on city_ascii and destination city or country
df_merged = pd.merge(df, lats_long, how='inner', left_on=['destination city'], right_on=['city_ascii'])
```

```python
df_merged['project date']
```

```python
# Replace Null project date with year, month, day
df_merged['project date'] = df_merged['project date'].fillna(pd.to_datetime(df_merged[['year','month','day']]))
```
