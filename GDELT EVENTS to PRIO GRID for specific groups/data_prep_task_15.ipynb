```python
import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_csv('goldstein_index_1990_to_2022_raw.csv')
```

```python
df
```

```python
#Rename COP, GOV, JUD, MIL, LEG, SPY to MajPol
df['Actor1Code'] = df['Actor1Code'].replace(['COP', 'GOV', 'JUD', 'MIL', 'LEG', 'SPY'], 'MajPol')
df['Actor2Code'] = df['Actor2Code'].replace(['COP', 'GOV', 'JUD', 'MIL', 'LEG', 'SPY'], 'MajPol')

#Rename INS, REB, SEP, OPP, IMG to MinPol
df['Actor1Code'] = df['Actor1Code'].replace(['INS', 'REB', 'SEP', 'OPP', 'IMG'], 'MinPol')
df['Actor2Code'] = df['Actor2Code'].replace(['INS', 'REB', 'SEP', 'OPP', 'IMG'], 'MinPol')

#Rename BUS, MNC to Bus
df['Actor1Code'] = df['Actor1Code'].replace(['BUS', 'MNC'], 'Bus')
df['Actor2Code'] = df['Actor2Code'].replace(['BUS', 'MNC'], 'Bus')

#Rename ELI to Eli
df['Actor1Code'] = df['Actor1Code'].replace(['ELI'], 'Eli')
df['Actor2Code'] = df['Actor2Code'].replace(['ELI'], 'Eli')

#Rename DEV, EDU, HLH, HRI, REF, IGO, NGO, NGM to CivSoc
df['Actor1Code'] = df['Actor1Code'].replace(['DEV', 'EDU', 'HLH', 'HRI', 'REF', 'IGO', 'NGO', 'NGM'], 'CivSoc')
df['Actor2Code'] = df['Actor2Code'].replace(['DEV', 'EDU', 'HLH', 'HRI', 'REF', 'IGO', 'NGO', 'NGM'], 'CivSoc')

#Rename AGR, ENV to Land
df['Actor1Code'] = df['Actor1Code'].replace(['AGR', 'ENV'], 'Land')
df['Actor2Code'] = df['Actor2Code'].replace(['AGR', 'ENV'], 'Land')

#Rename LAB, CVL to People
df['Actor1Code'] = df['Actor1Code'].replace(['LAB', 'CVL'], 'People')
df['Actor2Code'] = df['Actor2Code'].replace(['LAB', 'CVL'], 'People')

#Rename MED to Med
df['Actor1Code'] = df['Actor1Code'].replace(['MED'], 'Med')
df['Actor2Code'] = df['Actor2Code'].replace(['MED'], 'Med')
```

```python
df
```

```python
#Create different dataframes for as per Actor2Code
df_MajPol = df[df['Actor2Code'] == 'MajPol']
df_MinPol = df[df['Actor2Code'] == 'MinPol']
df_Bus = df[df['Actor2Code'] == 'Bus']
df_Eli = df[df['Actor2Code'] == 'Eli']
df_CivSoc = df[df['Actor2Code'] == 'CivSoc']
df_Land = df[df['Actor2Code'] == 'Land']
df_People = df[df['Actor2Code'] == 'People']
df_Med = df[df['Actor2Code'] == 'Med']
```

```python
#Pivot df to place Actor1Code as columns. Get mean of GoldsteinScale
df_MajPol_pivot = df_MajPol.pivot_table(index=['year', 'pg_lat', 'pg_long'], columns='Actor1Code', values='GoldsteinScale', aggfunc=['mean'])
df_MajPol_pivot = df_MajPol_pivot.reset_index().reset_index(drop=True)
df_MajPol_pivot.columns = ['_'.join(col).strip() for col in df_MajPol_pivot.columns.values]
df_MajPol_pivot.columns = df_MajPol_pivot.columns.str.rstrip('_')
df_MajPol_pivot

df_MinPol_pivot = df_MinPol.pivot_table(index=['year', 'pg_lat', 'pg_long'], columns='Actor1Code', values='GoldsteinScale', aggfunc=['mean'])
df_MinPol_pivot = df_MinPol_pivot.reset_index().reset_index(drop=True)
df_MinPol_pivot.columns = ['_'.join(col).strip() for col in df_MinPol_pivot.columns.values]
df_MinPol_pivot.columns = df_MinPol_pivot.columns.str.rstrip('_')
df_MinPol_pivot

df_Bus_pivot = df_Bus.pivot_table(index=['year', 'pg_lat', 'pg_long'], columns='Actor1Code', values='GoldsteinScale', aggfunc=['mean'])
df_Bus_pivot = df_Bus_pivot.reset_index().reset_index(drop=True)
df_Bus_pivot.columns = ['_'.join(col).strip() for col in df_Bus_pivot.columns.values]
df_Bus_pivot.columns = df_Bus_pivot.columns.str.rstrip('_')
df_Bus_pivot

df_Eli_pivot = df_Eli.pivot_table(index=['year', 'pg_lat', 'pg_long'], columns='Actor1Code', values='GoldsteinScale', aggfunc=['mean'])
df_Eli_pivot = df_Eli_pivot.reset_index().reset_index(drop=True)
df_Eli_pivot.columns = ['_'.join(col).strip() for col in df_Eli_pivot.columns.values]
df_Eli_pivot.columns = df_Eli_pivot.columns.str.rstrip('_')
df_Eli_pivot

df_CivSoc_pivot = df_CivSoc.pivot_table(index=['year', 'pg_lat', 'pg_long'], columns='Actor1Code', values='GoldsteinScale', aggfunc=['mean'])
df_CivSoc_pivot = df_CivSoc_pivot.reset_index().reset_index(drop=True)
df_CivSoc_pivot.columns = ['_'.join(col).strip() for col in df_CivSoc_pivot.columns.values]
df_CivSoc_pivot.columns = df_CivSoc_pivot.columns.str.rstrip('_')
df_CivSoc_pivot

df_Land_pivot = df_Land.pivot_table(index=['year', 'pg_lat', 'pg_long'], columns='Actor1Code', values='GoldsteinScale', aggfunc=['mean'])
df_Land_pivot = df_Land_pivot.reset_index().reset_index(drop=True)
df_Land_pivot.columns = ['_'.join(col).strip() for col in df_Land_pivot.columns.values]
df_Land_pivot.columns = df_Land_pivot.columns.str.rstrip('_')
df_Land_pivot

df_People_pivot = df_People.pivot_table(index=['year', 'pg_lat', 'pg_long'], columns='Actor1Code', values='GoldsteinScale', aggfunc=['mean'])
df_People_pivot = df_People_pivot.reset_index().reset_index(drop=True)
df_People_pivot.columns = ['_'.join(col).strip() for col in df_People_pivot.columns.values]
df_People_pivot.columns = df_People_pivot.columns.str.rstrip('_')
df_People_pivot

df_Med_pivot = df_Med.pivot_table(index=['year', 'pg_lat', 'pg_long'], columns='Actor1Code', values='GoldsteinScale', aggfunc=['mean'])
df_Med_pivot = df_Med_pivot.reset_index().reset_index(drop=True)
df_Med_pivot.columns = ['_'.join(col).strip() for col in df_Med_pivot.columns.values]
df_Med_pivot.columns = df_Med_pivot.columns.str.rstrip('_')
df_Med_pivot
```

```python
df_MajPol_pivot.drop(['mean_MajPol'], axis=1, inplace=True)
df_MinPol_pivot.drop(['mean_MinPol'], axis=1, inplace=True)
df_Bus_pivot.drop(['mean_Bus'], axis=1, inplace=True)
df_Eli_pivot.drop(['mean_Eli'], axis=1, inplace=True)
df_CivSoc_pivot.drop(['mean_CivSoc'], axis=1, inplace=True)
df_Land_pivot.drop(['mean_Land'], axis=1, inplace=True)
df_People_pivot.drop(['mean_People'], axis=1, inplace=True)
# df_Med_pivot.drop(['mean_Med'], axis=1, inplace=True) not present for obvious reason
```

```python
#Remove mean_CRM, mean_INT, mean_RAD, mean_UAF, mean_UIS from all dataframes
df_MajPol_pivot.drop(['mean_CRM', 'mean_INT', 'mean_RAD', 'mean_UAF', 'mean_UIS'], axis=1, inplace=True)
df_MinPol_pivot.drop(['mean_CRM', 'mean_INT', 'mean_RAD', 'mean_UAF', 'mean_UIS'], axis=1, inplace=True)
df_Bus_pivot.drop(['mean_CRM', 'mean_INT', 'mean_RAD', 'mean_UAF', 'mean_UIS'], axis=1, inplace=True)
df_Eli_pivot.drop(['mean_CRM', 'mean_INT', 'mean_RAD', 'mean_UAF', 'mean_UIS'], axis=1, inplace=True)
df_CivSoc_pivot.drop(['mean_CRM', 'mean_INT', 'mean_RAD', 'mean_UAF', 'mean_UIS'], axis=1, inplace=True)
df_Land_pivot.drop(['mean_CRM', 'mean_INT', 'mean_RAD', 'mean_UAF', 'mean_UIS'], axis=1, inplace=True)
df_People_pivot.drop(['mean_CRM', 'mean_INT', 'mean_RAD', 'mean_UAF', 'mean_UIS'], axis=1, inplace=True)
df_Med_pivot.drop(['mean_CRM', 'mean_INT', 'mean_RAD', 'mean_UAF', 'mean_UIS'], axis=1, inplace=True)
```

```python
#Remove mean_ from column names and add group name from dataframe to end
df_MajPol_pivot.columns = df_MajPol_pivot.columns.str.replace('mean_', '')
df_MajPol_pivot.columns = df_MajPol_pivot.columns + '_MajPol'

df_MinPol_pivot.columns = df_MinPol_pivot.columns.str.replace('mean_', '')
df_MinPol_pivot.columns = df_MinPol_pivot.columns + '_MinPol'

df_Bus_pivot.columns = df_Bus_pivot.columns.str.replace('mean_', '')
df_Bus_pivot.columns = df_Bus_pivot.columns + '_Bus'

df_Eli_pivot.columns = df_Eli_pivot.columns.str.replace('mean_', '')
df_Eli_pivot.columns = df_Eli_pivot.columns + '_Eli'

df_CivSoc_pivot.columns = df_CivSoc_pivot.columns.str.replace('mean_', '')
df_CivSoc_pivot.columns = df_CivSoc_pivot.columns + '_CivSoc'

df_Land_pivot.columns = df_Land_pivot.columns.str.replace('mean_', '')
df_Land_pivot.columns = df_Land_pivot.columns + '_Land'

df_People_pivot.columns = df_People_pivot.columns.str.replace('mean_', '')
df_People_pivot.columns = df_People_pivot.columns + '_People'

df_Med_pivot.columns = df_Med_pivot.columns.str.replace('mean_', '')
df_Med_pivot.columns = df_Med_pivot.columns + '_Med'

df_MajPol_pivot.rename(columns={'year_MajPol':'year', 'pg_lat_MajPol':'pg_lat', 'pg_long_MajPol':'pg_long'}, inplace=True)
df_MinPol_pivot.rename(columns={'year_MinPol':'year', 'pg_lat_MinPol':'pg_lat', 'pg_long_MinPol':'pg_long'}, inplace=True)
df_Bus_pivot.rename(columns={'year_Bus':'year', 'pg_lat_Bus':'pg_lat', 'pg_long_Bus':'pg_long'}, inplace=True)
df_Eli_pivot.rename(columns={'year_Eli':'year', 'pg_lat_Eli':'pg_lat', 'pg_long_Eli':'pg_long'}, inplace=True)
df_CivSoc_pivot.rename(columns={'year_CivSoc':'year', 'pg_lat_CivSoc':'pg_lat', 'pg_long_CivSoc':'pg_long'}, inplace=True)
df_Land_pivot.rename(columns={'year_Land':'year', 'pg_lat_Land':'pg_lat', 'pg_long_Land':'pg_long'}, inplace=True)
df_People_pivot.rename(columns={'year_People':'year', 'pg_lat_People':'pg_lat', 'pg_long_People':'pg_long'}, inplace=True)
df_Med_pivot.rename(columns={'year_Med':'year', 'pg_lat_Med':'pg_lat', 'pg_long_Med':'pg_long'}, inplace=True)
```

```python
prio_grid = pd.read_csv('../PRIO GRID spine.csv') #read in PRIO GRID spine file

#Create date_df with year column containing 1990 to 2022
date_df = pd.DataFrame()
date_df['year'] = range(1990, 2023)
date_df['key'] = 1
prio_grid['key'] = 1
base_df = pd.merge(date_df, prio_grid, on ='key').drop("key", 1)
base_df
```

```python
#Merge all dataframes with base_df to create one dataframe with all years, latitudes and longitudes
df_MajPol_pivot = pd.merge(base_df, df_MajPol_pivot, left_on = ['lat','lon','year'], right_on = ['pg_lat','pg_long','year'], how='left').drop(['pg_lat','pg_long'], 1)
df_MinPol_pivot = pd.merge(base_df, df_MinPol_pivot, left_on = ['lat','lon','year'], right_on = ['pg_lat','pg_long','year'], how='left').drop(['pg_lat','pg_long'], 1)
df_Bus_pivot = pd.merge(base_df, df_Bus_pivot, left_on = ['lat','lon','year'], right_on = ['pg_lat','pg_long','year'], how='left').drop(['pg_lat','pg_long'], 1)
df_Eli_pivot = pd.merge(base_df, df_Eli_pivot, left_on = ['lat','lon','year'], right_on = ['pg_lat','pg_long','year'], how='left').drop(['pg_lat','pg_long'], 1)
df_CivSoc_pivot = pd.merge(base_df, df_CivSoc_pivot, left_on = ['lat','lon','year'], right_on = ['pg_lat','pg_long','year'], how='left').drop(['pg_lat','pg_long'], 1)
df_Land_pivot = pd.merge(base_df, df_Land_pivot, left_on = ['lat','lon','year'], right_on = ['pg_lat','pg_long','year'], how='left').drop(['pg_lat','pg_long'], 1)
df_People_pivot = pd.merge(base_df, df_People_pivot, left_on = ['lat','lon','year'], right_on = ['pg_lat','pg_long','year'], how='left').drop(['pg_lat','pg_long'], 1)
df_Med_pivot = pd.merge(base_df, df_Med_pivot, left_on = ['lat','lon','year'], right_on = ['pg_lat','pg_long','year'], how='left').drop(['pg_lat','pg_long'], 1)

#Merge all dataframes into one dataframe
df = pd.merge(df_MajPol_pivot, df_MinPol_pivot, left_on = ['lat','lon','year', 'gid'], right_on = ['lat','lon','year','gid'], how='left')
df = pd.merge(df, df_Bus_pivot, left_on = ['lat','lon','year', 'gid'], right_on = ['lat','lon','year','gid'], how='left')
df = pd.merge(df, df_Eli_pivot, left_on = ['lat','lon','year', 'gid'], right_on = ['lat','lon','year','gid'], how='left')
df = pd.merge(df, df_CivSoc_pivot, left_on = ['lat','lon','year', 'gid'], right_on = ['lat','lon','year','gid'], how='left')
df = pd.merge(df, df_Land_pivot, left_on = ['lat','lon','year', 'gid'], right_on = ['lat','lon','year','gid'], how='left')
df = pd.merge(df, df_People_pivot, left_on = ['lat','lon','year', 'gid'], right_on = ['lat','lon','year','gid'], how='left')
df = pd.merge(df, df_Med_pivot, left_on = ['lat','lon','year', 'gid'], right_on = ['lat','lon','year','gid'], how='left')
```

```python
#Drop lon and lat columns
df = df.drop(['lon','lat'], 1)

#Rename gid to PRIO GRID
df.rename(columns={'gid':'PRIO GRID'}, inplace=True)
```
