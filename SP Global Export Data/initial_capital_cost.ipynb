This code wrangles the initial cost estimate from SPG dataset

Study 1 would be the most recent, 2 the second most recent, 3 the third, etc.

The ICC will change from study to study as new assumptions are made when providing the cost. Study 1 is the most recent study available and more relevant, Study 1 is recommended. 


```python
import numpy as np
import pandas as pd
```

```python
def process(path):
    tmp = pd.read_csv(path)
    print(tmp.shape)
    tmp = tmp.drop([0,2,3], axis=0)
    tmp.head()
    df = tmp.iloc[1:,:]
    df.columns = tmp.iloc[0]
    return df
```

```python
df_1 = process('./input/initial_capital_cost_1.csv')
df_2 = process('./input/initial_capital_cost_2.csv')
df_3 = process('./input/initial_capital_cost_3.csv')

df_1.head()
```

```python
df = pd.concat([df_1,df_2,df_3],axis=0, keys = df_1.columns,ignore_index = True)
print(df.shape)
```
